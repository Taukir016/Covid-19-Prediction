# Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline

import plotly  #plotly graph objects are a high-level interface 
import plotly.express as px
import plotly.graph_objects as go
#plt.rcParams['figure.figsize']=17,8
import cufflinks as cf  #Cufflinks is another library that connects the Pandas data frame with Plotly enabling
import plotly.offline as pyo 
from plotly.offline import init_notebook_mode,plot,iplot

import folium
```

```python
pyo.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df=pd.read_excel("Covid cases in India.xlsx")
```

```python
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df
```

```python
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
df
```

```python
total_cases_overall=df['Total Cases'].sum()
print('The Total Number of Cases till now in India is :',total_cases_overall)
```

```python
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
```

```python
df
```

```python
Total_Active_Cases=df.groupby('Name of State / UT')['Active Cases'].sum().sort_values(ascending=False).to_frame()
```

```python
Indian_cord = pd.read_excel('Indian Coordinates.xlsx')
```

```python
Indian_cord
```

```python
df_full=pd.merge(Indian_cord,df,on='Name of State / UT')
```

```python
df_full
```

```python
map=folium.Map(location=[20,90],zoom_start=4,tiles='Stamenterrain')

for lat,long,value, name in zip(df_full['Latitude'],df_full['Longitude'],df_full['Total Cases'],df_full['Name of State / UT']):
    folium.CircleMarker([lat,long],radium=value*0.8,popup=('<strong>State</strong>: '+str(name).capitalize()+'<br>''<strong>Total Cases</strong>: ' + str(value)+ '<br>'),color='red',fill_color='orange',fill_opacity=0.3).add_to(map)
```

```python
bdb_India=pd.read_excel('per_day_cases.xlsx',parse_dates=True,sheet_name="India")
bdb_Italy=pd.read_excel('per_day_cases.xlsx',parse_dates=True,sheet_name="Italy")
bdb_Korea=pd.read_excel('per_day_cases.xlsx',parse_dates=True,sheet_name="Korea")
bdb_Wuhan=pd.read_excel('per_day_cases.xlsx',parse_dates=True,sheet_name="Wuhan")
```

```python
bdb_India
```

```python
fig=plt.figure(figsize=(10,5),dpi=200)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed Cases in India")
axes.bar(bdb_India["Date"],bdb_India["Total Cases"],color='Blue')
fig=px.bar(bdb_India,x="Date",y="Total Cases", color="Total Cases",title="Confirmed Cases in India")
fig.show()
```

```python
fig=px.bar(bdb_Italy,x="Date",y="Total Cases",color='Total Cases',title='Confirmed cases in Italy')
fig.show()
fig=px.bar(bdb_Korea,x="Date",y="Total Cases",color='Total Cases',title='Confirmed cases in Korea')
fig.show()
fig=px.bar(bdb_Wuhan,x="Date",y="Total Cases",color='Total Cases',title='Confirmed cases in Wuhan')
fig.show()
```

```python
#Subplots using Bar Graph
```

```python
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(
    rows=2,cols=2,
    specs=[[{"secondary_y":True},{"secondary_y":True}],[{"secondary_y":True},{"secondary_y":True}]],
    subplot_titles=("S.Korea","Italy","India","Wuhan"))

fig.add_trace(go.Bar(x=bdb_Korea['Date'],y=bdb_Korea['Total Cases'],
                    marker=dict(color=bdb_Korea['Total Cases'],coloraxis="coloraxis")),1,1)

fig.add_trace(go.Bar(x=bdb_Italy['Date'],y=bdb_Italy['Total Cases'],
                    marker=dict(color=bdb_Italy['Total Cases'],coloraxis="coloraxis")),1,2)

fig.add_trace(go.Bar(x=bdb_India['Date'],y=bdb_India['Total Cases'],
                    marker=dict(color=bdb_India['Total Cases'],coloraxis="coloraxis")),2,1)

fig.add_trace(go.Bar(x=bdb_Wuhan['Date'],y=bdb_Wuhan['Total Cases'],
                    marker=dict(color=bdb_Wuhan['Total Cases'],coloraxis="coloraxis")),2,2)



fig.update_layout(coloraxis=dict(colorscale='Bluered_r'),showlegend=False,title_text="Total Cases in 4 Countries")

fig.update_layout(plot_bgcolor='rgb(230,230,230)')
```
